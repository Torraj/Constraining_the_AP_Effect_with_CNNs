```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as sp
```

```python
O_r_fid = 9*10**(-5)  # fiducial radiation density parameter
O_m_fid = 0.31        # fiducial matter density parameter
z = 0.5               # calculating distances at a fixed redshift
H_0_fid = 70          # fiducial Hubble constant in km/s/Mpc

# Define function for inverse of Hubble parameter with arguments of z and cosmological parameters
def inv_H(z, O_r, O_m, H_0):
    return 1 / (H_0 * np.sqrt(O_r*(1+z)**4 + O_m*(1+z)**3 + 1 - O_m))

# Function computes ratio of perpendicular/parallel distances
def S_ratio(O_m):
    S_perp = sp.quad(inv_H, 0, z, args = (O_r_fid, O_m, H_0_fid))[0]
    S_para = inv_H(z, O_r_fid, O_m_fid, H_0_fid)
    return S_perp/S_para
```

```python
# Ratio of perpendicular/parallel distances for fiducial mass density parameter
S_ratio_fid = S_ratio(O_m_fid)

# Number of mass densities to test
M_num = 101
# Varying O_m from 0.29 to 0.33
min_O_m = 0.29
max_O_m = 0.33
O_m_range = np.linspace(min_O_m, max_O_m, M_num)

# Define array to hold possible AP stretches for different O_m values
# AP stretch defined as test ratio over fiducial ratio
AP_stretch = np.zeros(M_num)
for i in range(M_num):
    S_ratio_test = S_ratio(O_m_range[i])
    AP_stretch[i] = S_ratio_test/S_ratio_fid
```

Now we have to figure out what the Hubble parameter is when dark energy is allowed to vary. We should have:

$$H(z) = H_0 \sqrt{\Omega_r (1+z)^4 + \Omega_m (1+z)^3 + (1-\Omega_m) (1+z)^{3(1+w_0+w_a)} e^{-3w_a(\frac{z}{z+1})}}$$

```python
# Define function for inverse of Hubble parameter with arguments of z and time-varying dark energy
def inv_H_DE(z, O_r, O_m, H_0, w_0, w_a):
    return 1 / (H_0 * np.sqrt(O_r*(1+z)**4 + O_m*(1+z)**3 + (1 - O_m) * (1+z)**(3*(1 + w_0 + w_a)) * np.exp(-3*w_a*(z/(z+1)))))

# Function computes ratio of perpendicular/parallel distances, this time allowing for time-varying dark energy
def S_ratio_DE(O_m, w_0, w_a):
    S_perp = sp.quad(inv_H_DE, 0, z, args = (O_r_fid, O_m, H_0_fid, w_0, w_a))[0]
    S_para = inv_H(z, O_r_fid, O_m_fid, H_0_fid, w_0, w_a)
    return S_perp/S_para
```

```python
# Ratio of perpendicular/parallel distances for fiducial mass density parameter and dark energy w0wa
w_0_fid = -1
w_a_fid = 0
S_ratio_fid = S_ratio(O_m_fid, w_0_fid, w_a_fid)

# Number of varying values of w_0 and w_a to test
w_num = 13
# Varying w_0 from -1 to 0
min_w_0 = -1
max_w_0 = 0
w_0_range = np.linspace(min_w_0, max_w_0, w_num)

# Varying w_a from -2 to 1
min_w_a = -2
max_w_a = 1
w_a_range = np.linspace(min_w_a, max_w_a, w_num)

# Define array to hold possible AP stretches for different w0wa values
# AP stretch defined as test ratio over fiducial ratio
AP_stretch = np.zeros((w_num, w_num))
for i in range(w_num):
    for j in range(w_num):
        S_ratio_test = S_ratio(O_m_fid, w_0_range[i], w_a_range[j])
        AP_stretch[i][j] = S_ratio_test/S_ratio_fid
```
